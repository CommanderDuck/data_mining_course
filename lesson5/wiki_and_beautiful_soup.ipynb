```python
req = requests.get('https://ru.wikipedia.org/wiki/День_сурка_(фильм)')
req.status_code
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
#поиск списка ссылок, удовлетворяющих нашим условиям
def link_search(soup):
    linkList = []
    for link in soup.findAll('a'):
        linkText = link.get('href')
        #список условий может быть другим, если вы ищете другого типа ссылки
        if linkText and 'http' in linkText and 'wikipedia' in linkText:
            linkList.append(linkText)
    return linkList
```

```python
link_search(soup)[2]
```

```python
curLink = 'https://ru.wikipedia.org/wiki/День_сурка_(фильм)')
for i in range(5):
    req = requests.get(curLink)
    soup = BeautifulSoup(req.text, 'html.parser')
    #можно менять индекс в зависимости от номера ссылки который вы желаете посмотреть
    curLink = link_search(soup)[0]
```
