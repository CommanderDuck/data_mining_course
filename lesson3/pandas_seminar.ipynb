Это могут быть совершенно случайные пропуски (например, когда данные плохо сохранились)  или пропуски, которые подчиняются определенной закономерности (например, у пассажиров третьего класса не было личных мест, следовательно, их номера мест не будут записаны). Также, могу предположить, что разные методы лучше работают с разными заменами средним, то способ, каким способом вы собираетесь использовать эти данные дальше.

Способ 0: Выкинуть все объекты с пропусками. 
Если у вас достаточно большая выборка и если вы думаете, что пропуски имеют случайный характер, думаю, можно использовать этот метод.

Достоинства: не вносит никаких неточностей в данные, просто сделать.

Недостатки: теряем часть выборки.Например, в нашей выборке, выкидывая все nan, мы теряем больше 60% не очень уж большого датасета, так что этот метод кажется неудачным. 

Способ 1: Заменить пропуски на какое константное значение, например на среднее или на константу.
    
Достоинства: позволяет работать с объектами, у которых есть не все данные.

Недостатки: ухудшает и делает менее точными разные статистические показатели.

В качестве примера заменим в колонке с возрастом пропущенные значения на средние

Способ 3: Попытаться понять природу данных и логически восстановить (полностью или частично, значения признаков).

Достоинства: те же, что и у предыдущего метода + можно стать точнее
Недостатки: не всегда есть такая возможность + можно угадать неправильно
    
Например, в нашем датасете, можно заметить, что почти у всех пассажиров третьего класса класс не указан и дать всем один номер каюты. (не уверена, что это хорошее решение в данном случае)

```python
def gen_uncertain_data():
    x = np.linspace(0, 30, 100)
    y = np.sin(x/6*np.pi) + np.random.normal(0, 0.02, size=x.shape)
    error = np.random.normal(0.1, 0.02, size=y.shape)
    return x, y, error
```

```python
import matplotlib.pyplot as plt
import seaborn

x,y, error = gen_uncertain_data()
plt.plot(x, y, color='blue')
plt.fill_between(x, y-error, y+error, facecolor='green')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
